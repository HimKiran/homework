```python
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

# Calculate Streamfunction

Functions modified based on Susan's `gork.f`.

explicit scheme + forward difference.

```python
def cal_streamfun(L=2000e3, R=2e6, nrelax=500):

    a=L          # domain length in x-axis
    b=L          # domain length in y-axis
    ws=1.0       # windstress coefficient
    H=4000       # layer depth
    delta=40     # bottom Ekman layer typical length
    rho=1000     # density of fluid
    #f=1e-4      # Coriolis parameter
    
    f=np.sqrt(9.8*H)/R # Calculate f by Rossby Radius 
    beta=1.6e-11       # partial f partial y
    
    Be=2*beta*H/(f*delta)  # beta term
    gamma=-2*np.pi*ws/(rho*delta*f*a) # wind stress term
    
    npts=41 # number of x grid
    d=a/(npts-1.0) # grid size
    mpts=int(b/d+1) # number of y grid
    
    # allocation for streamfunction
    psi = np.zeros([npts, mpts])
    # create x, y grid
    x, y = np.meshgrid(np.arange(0, a+d, d), np.arange(0, b+d, d))
    
    for k in range(nrelax):
        for i in range(1, npts-1):
            for j in range(1,mpts-1):
                d_psi = (psi[i+1,j]-psi[i-1,j])*d/8.
                d2_psi = (psi[i+1,j]+psi[i-1,j]+psi[i,j+1]+psi[i,j-1])/4.
                #wind_stress = np.sin(np.pi*(i-1.)/(npts-1.))*d*d/4. <--- gork.f              
                wind_stress = 1 # <--- inside the circle
                if ((x[i, j]-a/2)**2+(y[i, j]-a/2)**2 > (a/3)**2):
                    wind_stress = 0 # <--- outside the circle
                psi[i, j] = -gamma*wind_stress + d2_psi + Be*d_psi
                
    return x, y, psi.T # Python is row-major, FORTRAN is col-major
```

# Flow changes with domain size ( L )

```python
# L = 20e3 case
x1, y1, psi1 = cal_streamfun(L=20e3, R=2e6, nrelax=500)
# L = 200e3 case
x2, y2, psi2 = cal_streamfun(L=200e3, R=2e6, nrelax=500)
# L = 2000e3 case
x3, y3, psi3 = cal_streamfun(L=2000e3, R=2e6, nrelax=500)
```

```python
fig=plt.figure(figsize=(17, 5))
ax1=plt.subplot2grid((1, 3), (0, 0), colspan=1, rowspan=1)
ax2=plt.subplot2grid((1, 3), (0, 1), colspan=1, rowspan=1)
ax3=plt.subplot2grid((1, 3), (0, 2), colspan=1, rowspan=1)
L=20e3 # Length of domain
ax1.contour(x1, y1, psi1, 10, colors=('k',), linewidths=2)
circle=plt.Circle((L/2, L/2), L/3, edgecolor='k', facecolor='r', alpha=0.4)
ax1.add_artist(circle); ax1.grid()
ax1.set_title('L=20e3, R=L, nrelax=500', fontsize=14)
L=200e3 # Length of domain
ax2.contour(x2, y2, psi2, 10, colors=('k',), linewidths=2)
circle=plt.Circle((L/2, L/2), L/3, edgecolor='k', facecolor='r', alpha=0.4)
ax2.add_artist(circle); ax2.grid()
ax2.set_title('L=200e3, R=L, nrelax=500', fontsize=14)
L=2000e3 # Length of domain
ax3.contour(x3, y3, psi3, 10, colors=('k',), linewidths=2)
circle=plt.Circle((L/2, L/2), L/3, edgecolor='k', facecolor='r', alpha=0.4)
ax3.add_artist(circle); ax3.grid()
ax3.set_title('L=2000e3, R=L, nrelax=500', fontsize=14)
```
