## Gauss quadrature

From <a href='http://en.wikipedia.org/wiki/Gaussian_quadrature'>**Wikipedia - Gaussian quadrature**</a>:
> n-point Gaussian quadrature is a numerical quadrature rule constructed to yield an exact result for polynomials of degree $\mathrm{2n-1}$ or less by a suitable choice of the points $x_i$ and weights $\omega_i$ for $n = 1, 2, \ldots, n$.

> \begin{equation*}
    \int_{-1}^1 f(x)\,dx \approx \sum_{i=1}^n w_i f(x_i)
  \end{equation*}

> Gaussian quadrature will only produce accurate results if the function $f(x)$ is well approximated by a polynomial function within the range $\mathrm{[−1,\ 1]}$. For the simplest integration problem, the associated polynomials are **Legendre polynomials** $P_n(x)$, and the method is usually known as Gauss–Legendre quadrature. With the n-th polynomial normalized to give $P_n(1)=1$, the i-th Gauss node, $x_i$, is the i-th root of $P_n$; its weight is:

> \begin{equation*}
    w_i = \frac{2}{\left( 1-x_i^2 \right) [P'_n(x_i)]^2}
  \end{equation*}

Here we work in the range of $x\ \epsilon\ \mathrm{[0,\ 20]}$, so we will <a href='http://en.wikipedia.org/wiki/Gaussian_quadrature#Change_of_interval'>**change the interval**</a> in the following way:

> \begin{equation*}
    \int_a^b f(x)\,dx \approx \frac{b-a}{2} \sum_{i=1}^n w_i f\left(\frac{b-a}{2}z_i + \frac{a+b}{2}\right)
  \end{equation*}

By looking at our function, Gauss-Legendre method works. 

Here I use `numpy.polynomial.legendre.leggauss` calculates the root of Legendre polynomials $x_i$ and weights $\omega_i$. (I can also use Table A-1 in the handout, but I don't want to type).

```python
def Gauss_Legendre(fun_name, a, b, n):
    '''
    Gauss-Legendre integral
    ============================================
    Input:
        fun_name: name of user defined function
        a, b: range of definite integral
        n: order of Legendre polynomials
    ============================================
    '''
    # calculate roots and weights
    x, w = np.polynomial.legendre.leggauss(n)
    dx = (b+a)/2*np.ones(x.shape)+((b-a)/2)*x
    # validate user defined function
    df=locals()['fun_name'](dx)
    result=((b-a)/2)*np.sum(w*df)
    return dx, df, result
```

## Spectral truncation to resolution

According to Warner's L1-L4 on p49 (also in Laprise, 1992), resolution of spectral model can be defined as:
$$
    L_1 = \frac{2\pi a}{3K+1} \qquad L_2 = \frac{\pi a}{K} \qquad L_3 = \frac{2\sqrt{\pi}a}{K+1} \qquad L_4 = \frac{\sqrt{2}\pi a}{K}
$$
So here I calculated the corresponds resolution for some of the truncations above:

```python
def trunc_res(K):
    K=np.array(K)
    a =6378.1 #km
    return 2.*np.pi*a/(3.*K+1.), np.pi*a/K, 2.*np.sqrt(np.pi)*a/(K+1.), np.sqrt(2.*np.pi*a)/K
```

```python
GFS = [24, 190, 574, 1534]; ECMWF = [159, 255, 319, 639, 1279]
GFS_res = trunc_res(GFS); EC_res = trunc_res(ECMWF)
print('GFS spectral resolution\n====================================================================================')
for i in range(4):
    print('T'+str(GFS[i])+' ---> L1: {}, L2: {}, L3: {}, L4: {}'.format(GFS_res[0][i], GFS_res[1][i], GFS_res[2][i], GFS_res[3][i]))
print('\nECMWF spectral resolution\n====================================================================================')
for i in range(5):
    print('T'+str(ECMWF[i])+' ---> L1: {}, L2: {}, L3: {}, L4: {}'.format(EC_res[0][i], EC_res[1][i], EC_res[2][i], EC_res[3][i]))
```

# Task III: Coiffier Eq. 4.22

Coiffier eq.(4.22) defines fully normalized associated Legendre function:

> \begin{equation*}
    P_n^m\left( \mu \right) = \sqrt{\left( 2n+1 \right)\frac{\left( n-m \right)!}{\left( n+m \right)!}} \frac{\left( 1-\mu^2 \right)^{\frac{m}{2}}}{2^nn!}\frac{d^{n+m}}{d \mu^{n+m}}\left( \mu^2-1 \right)^n
  \end{equation*}
  
And we know that, part of eq.(4.22) is the conventional associated Legendre functions:

\begin{equation*}
    P_n^m\left( \mu \right) = \frac{\left( 1-\mu^2 \right)^{\frac{m}{2}}}{2^nn!}\frac{d^{n+m}}{d \mu^{n+m}}\left( \mu^2-1 \right)^n
  \end{equation*}
  
But we cannot use equtions above due to the instability generated by high-order derivative term. 

According to `Abramowitz and Stegun (1972)`, associated Legendre functions has recursive form as:

  \begin{equation*}
    P_{n+1}^0\left( \mu \right) = (2n+1)xP_n^0\left( \mu \right) - nP_{n-1}^0\left( \mu \right) \\
    P_{n}^{n}\left( \mu \right) = (2n-1)P_1^1\left( \mu \right)P_{n-1}^{n-1}\left( \mu \right) \\
    P_n^m\left( \mu \right) = P_{n-2}^m\left( \mu \right) + (2n-1)P_1^1\left( \mu \right)P_{n-1}^{m-1}\left( \mu \right)
  \end{equation*}
  
With initial conditions for $P_0^0\left( \mu \right)$, $P_1^0\left( \mu \right)$, etc.

This is the algorithm I applied, more details please see:

* `Abramowitz, M. and I. A. Stegun, 1972. Handbook of Mathematical Functions. Dover Publications, Inc., New York, pp. 1046.`
* <a href='http://mitgcm.org/~mlosch/geoidcookbook/node11.html'>**MITgcm Guidebook, 3.7 Fully normalized associated Legendre functions**</a>

## Eq. 4.22

Here I defined `assoc_legendre` for associated and `norm_legendre` for normalized associated Legendre function group. 

```python
# factorial
from scipy.misc import factorial
# odd factorial
def odd_factorial(N):
    out=N
    while N >= 3:
        N -= 2
        out *= N
    return out
# Associated Legendre function
def assoc_legendre(m, n, x):
    '''
    Associated Legendre function
    =================================
    Input:
        n: order of Legendre function
        m: degree of association
        x: series -1< x < 1
    =================================
    '''
    if m < 0 or m > n:
        raise ValueError('Input error, m must be positive and smaller than n')
    # Initial cases.
    if (m == 0):
        p_n=1.0
    else:
        s=1
        if (m & 1):
            s=-1
        z=np.sqrt(1-x**2)
        p_n=s*odd_factorial(2*m - 1)*z**m
    if (m == n):
        return p_n
    # recursion.
    p_n_prev=p_n
    p_n=x*(2*m+1)*p_n
    if (n == (m+1)):
        return p_n
    # Iterate the recursion
    for i in range(m+2, n+1):
        result=(x*(2*i-1)*p_n-(i+m-1)*p_n_prev)/(i-m)
        p_n_prev=p_n
        p_n=result
    return result
# Normalized Legendre function
def norm_legendre(m, n, x):
    '''
    Normalized associated Legendre function
        supports NumPy arrary
    '''
    L=len(x)
    result=np.empty(L)
    for i in range(L):
        result[i]=assoc_legendre(m, n, x[i])
    result=result*np.sqrt((2*n+1)*factorial(n-m)/factorial(n+m))
    return result
```

# Task IV: Spectral methods solving PDE

We are now trying to solve *viscous Burgers' equation*, which is a simplified version of *Navier-Stokes equation*:

\begin{equation*}
    \frac{\partial T}{\partial t} + U_0 \frac{\partial T}{\partial x} = K \frac{\partial^2 T}{\partial x^2}
\end{equation*}

Here $U_0$ and $K$ are constants, so *Burgers' equation* is linearized.

It has a wave-like solution, in Fourier transform with wavenumber $m$, it is:
    $$
    T = \mathcal F\left( \hat{T}e^{i\xi x} \right), \qquad \xi = \frac{2m\pi}{L}
    $$

So we replace each $x$-derivative by $i \xi$ and replace $T$ by $\hat{T}$:

$$
    \hat{T_t} + i \xi U_0 \hat{T} = -\xi^2 K \hat{T}
$$
It is the ODE required by this homework.

The ODE above can be temporally discretizated by iterative methods (e.g. R.-K. method).

## Numerical scheme

Here I also provide a numerical solution for the ODE above:

* **Domain**: one-dimensional, 64 points, $x\ \epsilon\ [-\pi, \pi]$.
* **Initial condition**: $T(t=0) = 25^\circ C \cdot sin(x)^2$ for x < 0 only. 
* **Temporal discretization**: 4th order Runge-Kutta method.
* **Fourier transform**: one-dimensional discrete Fourier Transform (in `numpy.fft`).

Here I multiplying all the terms by an integrating factor $\exp(\xi^2 K t)$:
$$
    \quad \\
    e^{\xi^2 K t} \hat{T}_t + i \xi U_0 e^{\xi^2 K t} \hat{T} = -\xi^2 K e^{\xi^2 K t}\hat{T}
$$

Then define $\hat{U} = e^{\xi^2 K t} \hat{T}$:
$$
    \hat{U}_t = \xi^2 K \hat{U} + e^{\xi^2 K t} \hat{T}_t \\
    \hat{U}_t - \xi^2 K \hat{U}  + i \xi U_0 \hat{U} = -\xi^2 K \hat{U} \\
    \quad
$$

Therefore, the dissipative term has been canceled out, and I actually solving:
$$
    \hat{U}_t = -i \xi U_0 \hat{U}
$$

```python
# Set parameters
U0=0.5        # Advection coefficient
K=0.25        # Viscosity
m=64          # Number of grid points in 1-D
freq=1./m**2  # Time frequency 
total_time=3. # maximum time
# Create domain
x=np.arange(-m/2,m/2)*(2*np.pi/m)
# All timestep
max_timestep=int(round(total_time/freq))
# Initial condition in frequency domain
T=25*np.sin(x)**2 * (x<0.)
U=np.fft.fft(T) # <------------------------------------------------ FFT
# Allocate result
result=np.empty([max_timestep+1, len(x)]) # result
result[0, :]=T
time_count=np.empty([max_timestep+1]) # absulote time
time_count[0]=0
# Coefficients in semi-discretization equation
xi=np.array([range(0, m/2) + [0] + range(-m/2+1, 0)]) # create xi
C_adv = -1j*freq*xi*U0 # coefficient of right-hand term
EXP = np.exp(-freq*K*xi**2/2) # integrating factor
# Solve ODE
for i in range(1, max_timestep+1):
    C1 = C_adv*U
    C2 = C_adv*EXP*(U+C1/2.)
    C3 = C_adv*EXP*U+C2/2
    C4 = C_adv*(EXP*U+EXP*C3)
    U = EXP*U + (EXP*C1+2*EXP*(C2+C3)+C4)/6. # <------------- Runge-Kutta
    # collecting results
    time_count[i]=i*freq
    result[i, :]=np.squeeze(np.real(np.fft.ifft(U))) # <----- Inverse FFT
```

## Solution & discussions

```python
x_grid, t_grid = np.meshgrid(x, time_count[::500])
result_3Dplot = result[::500, :]
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.gca(projection='3d')
ax.plot_wireframe(x_grid, t_grid, result_3Dplot, rstride=2, cstride=100, color='k', linewidth=3, alpha=1)
ax.set_xlim3d(-np.pi, np.pi); ax.set_ylim3d(0, 3); ax.set_zlim3d(0, 25);
ax.set_xlabel('x grid', fontsize=12); ax.set_ylabel('time', fontsize=12); ax.set_zlabel('Temperature', fontsize=12)
ax.set_title('Solution in x-time frame', fontsize=14, fontweight='bold')
ax.view_init(20, -60)
```

```python
fig, ax = plt.subplots(3, 3, figsize=(10, 10))
fig.tight_layout()
t_select=[0, 0.25, 0.5, 0.75, 1, 1.5, 2, 2.5, 3]
for l in range(3):
    for m in range(3): 
        ax_temp=ax[l, m]; time_temp=t_select[3*l+m]
        ax_temp.plot(x, np.squeeze(result[time_count==time_temp, :]), color='k', linewidth=3)
        ax_temp.set_xlim([-np.pi, np.pi]); ax_temp.set_ylim([0,25])
        ax_temp.set_title('time='+str(time_temp), fontsize=12)
        ax_temp.grid()
```

From the figures above, we see that:
